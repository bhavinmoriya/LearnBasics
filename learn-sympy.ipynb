# Algebraic Manipulation


```python
# Expand expressions
expanded = expand((x + y)**3)
print(expanded)
# Result: x**3 + 3*x**2*y + 3*x*y**2 + y**3

# Factor expressions
factored = factor(x**2 - 1)
# Result: (x - 1)*(x + 1)

# Expand trigonometric expressions
trig_expanded = expand_trig(sin(x + y))
# Result: sin(x)*cos(y) + sin(y)*cos(x)
```

```python
# Basic simplification
expr = (x**2 - 1)/(x - 1)
simplified = simplify(expr)  # Result: x + 1

# Trigonometric simplification
trig_expr = sin(x)**2 + cos(x)**2
simplified_trig = trigsimp(trig_expr)  # Result: 1

# Rational function simplification
rational_expr = (x**2 + 2*x + 1)/(x + 1)
simplified_rational = simplify(rational_expr)  # Result: x + 1
```

```python
# Collect terms
expr = x*y + x*z + y*z
collected = collect(expr, x)  # Result: x*(y + z) + y*z

# Combine like terms
combined = x + 2*x + 3*x  # Automatically becomes 6*x

# Apart and together for rational functions
rational = (x + 1)/(x**2 + x)
apart_result = apart(rational, x)  # Partial fractions
together_result = together(apart_result)  # Combine fractions
```

# Calculus Operations

```python
# Basic derivatives
f = x**3 + 2*x**2 + x + 1
f_prime = diff(f, x)  # 3*x**2 + 4*x + 1

# Higher-order derivatives
f_double_prime = diff(f, x, 2)  # Second derivative
f_triple_prime = diff(f, x, 3)   # Third derivative

# Partial derivatives
f_xy = x**2 * y + y**3
df_dx = diff(f_xy, x)  # 2*x*y
df_dy = diff(f_xy, y)  # x**2 + 3*y**2
df_dxdy = diff(f_xy, x, y)  # Mixed partial: 2*x

# Chain rule examples
f = sin(x**2)
df_dx = diff(f, x)  # 2*x*cos(x**2)
```

```python
# Indefinite integrals
f = x**2
indefinite = integrate(f, x)  # x**3/3

# Definite integrals
definite = integrate(x**2, (x, 0, 1))  # 1/3

# Multiple integrals
double_integral = integrate(x*y, (x, 0, 1), (y, 0, 2))  # 1

# Improper integrals
improper = integrate(exp(-x), (x, 0, oo))  # 1

# Trigonometric integrals
trig_integral = integrate(sin(x)*cos(x), x)  # sin(x)**2/2
```

```python
# Basic limits
limit_result = limit(sin(x)/x, x, 0)  # 1

# Limits at infinity
limit_inf = limit(1/x, x, oo)  # 0

# One-sided limits
left_limit = limit(1/x, x, 0, '-')   # -oo
right_limit = limit(1/x, x, 0, '+')  # +oo

# L'Hôpital's rule (automatic)
indeterminate = limit((exp(x) - 1)/x, x, 0)  # 1
```

```python
# Taylor series
taylor = series(exp(x), x, 0, 6)  # 1 + x + x**2/2 + x**3/6 + x**4/24 + x**5/120 + O(x**6)

# Laurent series
laurent = series(sin(x)/x, x, 0, 6)

# Series around different points
series_pi = series(sin(x), x, pi, 4)

# Remove O() term
taylor_no_o = series(exp(x), x, 0, 6).removeO()
```

# Equation Solving

```python
# Linear equations
linear_eq = Eq(2*x + 3, 7)
solution = solve(linear_eq, x)  # [2]

# Quadratic equations
quadratic = x**2 - 4
solutions = solve(quadratic, x)  # [-2, 2]

# System of equations
eq1 = Eq(x + y, 5)
eq2 = Eq(x - y, 1)
system_solution = solve([eq1, eq2], [x, y])  # {x: 3, y: 2}

# Polynomial equations
poly = x**3 - 6*x**2 + 11*x - 6
poly_solutions = solve(poly, x)  # [1, 2, 3]
```

```python
# First-order ODEs
y = Function('y')
ode1 = Eq(y(x).diff(x), y(x))
ode1_solution = dsolve(ode1, y(x))  # y(x) = C1*exp(x)

# Second-order ODEs
ode2 = Eq(y(x).diff(x, 2) + y(x), 0)
ode2_solution = dsolve(ode2, y(x))  # y(x) = C1*cos(x) + C2*sin(x)

# With initial conditions
ode_ic = dsolve(ode1, y(x), ics={y(0): 1})  # y(x) = exp(x)

# System of ODEs
z = Function('z')
system_ode = [Eq(y(x).diff(x), z(x)), Eq(z(x).diff(x), -y(x))]
system_solution = dsolve(system_ode, [y(x), z(x)])
```

```python
# Equations with transcendental functions
transcendental = exp(x) - 2
trans_solution = solve(transcendental, x)  # [log(2)]

# Trigonometric equations
trig_eq = sin(x) - Rational(1, 2)
trig_solutions = solve(trig_eq, x)

# Using nsolve for numerical solutions
from sympy import nsolve
numerical_solution = nsolve(cos(x) - x, 0.5)
```

# Performance Tips

```python
# Use assumptions to speed up simplification
x_pos = Symbol('x', positive=True)
expr = sqrt(x_pos**2)  # Simplifies to x immediately

# Cache expensive computations
from sympy import cacheit
@cacheit
def expensive_calculation(n):
    return sum(i**2 for i in range(n))

# Use evalf for numerical evaluation
pi.evalf(50)  # 50 decimal places

# Avoid automatic simplification when not needed
expr = Add(x, x, evaluate=False)  # Keeps as x + x instead of 2*x
```

```python
# Clear caches when working with large expressions
from sympy.core.cache import clear_cache
clear_cache()

# Use generators for large sequences
def fibonacci_gen():
    a, b = 1, 1
    while True:
        yield a
        a, b = b, a + b
```
